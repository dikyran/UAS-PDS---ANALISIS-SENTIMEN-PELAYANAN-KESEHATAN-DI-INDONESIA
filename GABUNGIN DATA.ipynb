```python
import pandas as pd

# Daftar file CSV yang akan digabungkan
files = [
    r"C:\Users\DIKY\tweets-data\#BPJS_04-06-2024_10-39-13.csv",
r"C:\Users\DIKY\tweets-data\#BPJSKesehatan_08-06-2024_22-45-41.csv",
r"C:\Users\DIKY\tweets-data\#Dokter_04-06-2024_11-01-01.csv",
r"C:\Users\DIKY\tweets-data\#DokterIndonesia_08-06-2024_22-53-18.csv",
r"C:\Users\DIKY\tweets-data\#FasilitasKesehatan_08-06-2024_22-41-47.csv",
r"C:\Users\DIKY\tweets-data\#Imunisasi_04-06-2024_11-25-00.csv",
r"C:\Users\DIKY\tweets-data\#Jaminan_Kesehatan_Nasional_08-06-2024_21-35-33.csv",
r"C:\Users\DIKY\tweets-data\#Kemenkes_04-06-2024_12-20-11.csv",
r"C:\Users\DIKY\tweets-data\#Kementerian_Kesehatan_06-06-2024_11-48-53.csv",
r"C:\Users\DIKY\tweets-data\#Kesehatan_04-06-2024_12-00-00.csv",
r"C:\Users\DIKY\tweets-data\#LayananMedis_08-06-2024_22-50-08.csv",
r"C:\Users\DIKY\tweets-data\#PelayananKesehatan_04-06-2024_10-17-14.csv",
r"C:\Users\DIKY\tweets-data\#Perawat_06-06-2024_11-45-25.csv",
r"C:\Users\DIKY\tweets-data\#Puskesmas_04-06-2024_11-47-25.csv",
r"C:\Users\DIKY\tweets-data\#SehatIndonesia_04-06-2024_11-11-38.csv",
r"C:\Users\DIKY\tweets-data\#Vaksin_06-06-2024_12-30-36.csv"
]


# List untuk menyimpan DataFrame dari setiap file CSV
dataframes = []

# Loop melalui setiap file dan baca CSV ke DataFrame
for file in files:
    df = pd.read_csv(file)
    dataframes.append(df)

# Gabungkan semua DataFrame menjadi satu DataFrame
merged_df = pd.concat(dataframes, ignore_index=True)

# Simpan hasil penggabungan ke file CSV baru
output_file = r"C:\Users\DIKY\tweets-data\DATA FINAL.csv"
merged_df.to_csv(output_file, index=False)

print("File CSV berhasil digabungkan dan disimpan di:", output_file)
```
